```python
x_60=df.drop('esi',axis=1)
```

```python
y_60=df['esi']
```

```python
y_60=y_60.astype(int)
```

```python
y_60=y_60-1
```

```python
y_60.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_60, X_test_60, y_train_60, y_test_60 = train_test_split(x_60, y_60, test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
```

```python
initial_model_60 = xgb.XGBClassifier()
```

```python
initial_model_60.fit(X_train_60, y_train_60)
```

```python
y_pred_60 = initial_model_60.predict(X_test_60)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_60 = accuracy_score(y_test_60, y_pred_60)
print("Initial Model Accuracy:", accuracy_60)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Assuming y_true contains the true labels and y_pred contains the predicted labels
# y_true and y_pred should both be arrays with integer class labels (0 to N-1, where N is the number of classes)

# Calculate overall accuracy
overall_accuracy_60 = accuracy_score(y_test_60, y_pred_60)

# Calculate class-wise accuracy
class_wise_accuracy_60 = []
for class_label in range(5):  # Iterate over each class label
    class_indices = (y_test_60 == class_label)  # Get indices where the true label is equal to the current class label
    class_true = y_test_60[class_indices]  # True labels for the current class
    class_pred = y_pred_60[class_indices]  # Predicted labels for the current class

    # Calculate accuracy for the current class
    class_accuracy = accuracy_score(class_true, class_pred)
    class_wise_accuracy_60.append(class_accuracy)

# Print overall accuracy and class-wise accuracy
print("Overall Accuracy:", overall_accuracy_60)
for class_label, accuracy in enumerate(class_wise_accuracy_60):
    print(f"Class {class_label} Accuracy:", accuracy)

# Optionally, you can print a classification report for more detailed metrics
report_60 = classification_report(y_test_60, y_pred_60)
print(report_60)
```
